```python
#percentage of missing values in the bmi column
print(f"Percentage of missing values in 'bmi' column: {stroke['bmi'].isna().mean() * 100:.2f}%")
```

```python
#used imputation to fill the missing values in the bmi column with the mean of the column
stroke['bmi'] = stroke['bmi'].fillna(stroke['bmi'].mean())
```
