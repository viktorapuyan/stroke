Data Preprocessing

```python
stroke.isna().sum()
```

```python
#percentage of missing values in the bmi column
print(f"Percentage of missing values in 'bmi' column: {stroke['bmi'].isna().mean() * 100:.2f}%")
```

```python
#used imputation to fill the missing values in the bmi column with the mean of the column
stroke['bmi'] = stroke['bmi'].fillna(stroke['bmi'].mean())
```

```python
#check if missing values have been filled
stroke.isna().sum()
```

```python
import pandas as pd, numpy as np
stroke['bmi'] = pd.to_numeric(stroke['bmi'], errors='coerce')

# stroke prevalence
print("stroke prevalence:", stroke['stroke'].mean())

# stroke rate by category
for col in ['work_type','ever_married','Residence_type']:
    tab = stroke.groupby(col)['stroke'].agg(['count','sum'])
    tab['rate'] = tab['sum'] / tab['count']
    print("\n", col)
    print(tab.sort_values('rate', ascending=False))
```

```python
import scipy.stats as stats
for col in ['work_type','ever_married','Residence_type']:
    ct = pd.crosstab(stroke[col].fillna('Missing'), stroke['stroke'])
    chi2, p, dof, ex = stats.chi2_contingency(ct)
    print(col, "chi2 p-value:", p)
```
